```python
data['Licensed'] = data['Licensed'].replace({True: 1, False: 0})
data['Official Video'] = data['Official Video'].replace({True: 1, False: 0})

# Boolean bir ifade eğer T veya F olarak tutuluyorsa büyük verilerde yorucu bir işlem olabilir. Bundan dolayı
# bu değerleri sayısal bir ifadeye dönüştürmek gerekmektedir.
```

```python
import numpy as np

# Sayısal sütunları seçme
numeric_columns = data.select_dtypes(include=[np.number]).columns

# Aykırı değerleri tespit etme
def detect_outliers(df, columns):
    outliers = pd.DataFrame()
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[column] = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    return outliers

# Aykırı değerleri tespit et
outliers = detect_outliers(data, numeric_columns)

# Aykırı değerlerin özetini görüntüleme
print(outliers)
```

```python
import numpy as np

# Sayısal sütunları seçme
numeric_columns = data.select_dtypes(include=[np.number]).columns

# Aykırı değerleri temizleme (IQR yöntemi kullanarak)
Q1 = data[numeric_columns].quantile(0.25)
Q3 = data[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Alt ve üst sınırlar
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerlerin dışında kalan verileri seçme
cleaned_data = data[~((data[numeric_columns] < lower_bound) | (data[numeric_columns] > upper_bound)).any(axis=1)]

# Orijinal ve temizlenmiş veri setinin boyutlarını kontrol etme
print(f'Orijinal veri seti {data.shape[0]} satıra ve {data.shape[1]} sütuna sahiptir.')
print(f'Temizlenmiş veri seti {cleaned_data.shape[0]} satıra ve {cleaned_data.shape[1]} sütuna sahiptir.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from math import pi

# En yüksek Stream değerine sahip 10 şarkıyı seçme
top_songs = cleaned_data.sort_values('Stream', ascending=False).head(10)

# İncelenecek metrikleri seçme ve bu metriklerin yüzdelik hale getirilmesi
şarkı_metrikleri = top_songs[['Title', 'Energy', 'Danceability', 'Acousticness', 'Loudness', 'Speechiness', 'Instrumentalness', 'Liveness', 'Valence']].copy()
şarkı_metrikleri.loc[:, ['Energy', 'Danceability', 'Acousticness', 'Speechiness', 'Instrumentalness', 'Liveness', 'Valence']] *= 100

# Sayısal değerleri virgülden sonra iki basamak olacak şekilde yuvarlama
for col in ['Energy', 'Danceability', 'Acousticness', 'Loudness', 'Speechiness', 'Instrumentalness', 'Liveness', 'Valence']:
    şarkı_metrikleri[col] = şarkı_metrikleri[col].map(lambda x: round(x, 2))

# Radar Chart için veri hazırlama
categories = list(şarkı_metrikleri.columns[1:])
N = len(categories)

# Şarkıları tek tek ekleme
plt.figure(figsize=(10, 8))

for i, row in şarkı_metrikleri.iterrows():
    values = row.drop('Title').values.flatten().tolist()
    values += values[:1]  # Değerleri kapatmak için tekrar ilk değeri ekliyoruz
    
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
    
    ax = plt.subplot(111, polar=True)
    
    plt.xticks(angles[:-1], categories, color='grey', size=8)
    
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=row['Title'])
    ax.fill(angles, values, alpha=0.4)
    
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```
